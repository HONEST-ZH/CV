### 图像拆分通道 cv2.split()
函数 cv2.split() 将 3 通道 BGR 彩色图像分离为 **B、G、R** 单通道图像。
### 图像通道的合并 cv2.merge()
函数 cv2.split() 将3个单通道合并成一个 3通道图像。

```python
    import cv2
    import numpy as np
    
    img1 = cv2.imread("image/LenaNoise.png", flags=1)  # flags=1 读取彩色图像
    cv2.imshow("LenaNoise.png", img1) 

    cv2.waitKey(0)
    cv2.destroyAllWindows()  # 释放所有窗口
```

### 注意格式：
1. cv2.imread( ) 所读取的图像格式：每个像素为[B,G,R]的形式
2. cv2.imshow( ) 显示图像：按图像中每个像素为[B,G,R]的规则将图像显示出来

```python
    # BGR 通道拆分
    ch1, ch2, ch3 = cv2.split(img1)  # 拆分为 BGR 独立通道
    #cv2.imshow("rImg", rImg)  # 直接显示红色分量 rImg 显示为灰度图像
    print(img1.shape,ch1.shape, ch2.shape, ch3.shape)
    cv2.imshow("ch1", ch1) 
    cv2.imshow("ch2", ch2) 
    cv2.imshow("ch3", ch3) 
    cv2.waitKey(0)
    cv2.destroyAllWindows()  # 释放所有窗口
```

```python
    mergeImg = cv2.merge([ch1, ch2, ch3])
    cv2.imshow("mergeImg", mergeImg) 
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
```
